```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

畫函數的圖形。

$f(x)=2x^{3}+3x^{2}+x+180x\cdot sin(0.8x)+5$

```python
x = np.linspace(-10, 10, 200) #from -10 to 10 draw 200 dots
y = 2*x**3 + 3*x**2 + x + 180*x*np.sin(0.8*x) + 5
```

```python
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(x,y)
```

```python
x = np.linspace(0.5, 2, 200)
```
