```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

$\begin{cases}
-4 \sqrt{1-\frac{x^2}{5^2}}, \quad \text{if }-5\leq x \leq 5\\
-2x+10, \quad \text{if }0\leq x \leq 5\\
2\cdot (x+5), \quad \text{if }-5\leq x \leq 0\\
\end{cases}$

```python
x1 = np.linspace(-5,5,100)
y1 = -4*np.sqrt(1-x1*x1/25)

x2 = np.linspace(0,5,100)
y2 = -2*x2+10

x3 = np.linspace(-5,0,100)
y3 = 2*(x3+5)

x4 = np.linspace(0,5,100)
```

```python
plt.plot(x1,y1,'r')
plt.plot(x2,y2,'g')
plt.plot(x3,y3,'b')
plt.grid()
plt.show()
```


畫函數的圖形。

$f(x)=2x^{3}+3x^{2}+x+180x\cdot sin(0.8x)+5$

```python
x = np.linspace(-10, 10, 200) #from -10 to 10 draw 200 dots
y = 2*x**3 + 3*x**2 + x + 180*x*np.sin(0.8*x) + 5
```

```python
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(x,y) #plot graph
plt.show()
```

```python
x = np.linspace(0.5, 2, 200)
y = np.log(x)
```

```python
plt.plot(x,y)
plt.show()
```
