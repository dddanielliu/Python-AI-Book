```python
# 使用 pipeline

from transformers import pipeline
```

### 2. 瞬間打造生成文字的機器

```python
generator_en = pipeline("text-generation")
```

```python
[text_en] = generator_en("I am a student who loves programming.")
```

```python
text_en['generated_text']
```

### 3. 情意分析

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier("This device was actually even better than number two, but still had some huge problems with Asus' quality assurance.")
```

### 4. Zero-Shot 文字分類!!

```python
classifier = pipeline("zero-shot-classification")
```

```python
classifier(
    "ASUS ROG STRIX G15 (G513) has excellent specs and characteristics. Its powerful hardware copes with the latest games in FHD resolution. It’s quite comfortable to run even the latest games on the ASUS ZEPHYRUS G15 GA501I.",
    candidate_labels=["education", "tech", "sport"],
)
```

### 5. 指定 model

```python
generator = pipeline("text-generation", model="ckiplab/gpt2-base-chinese")
```

```python
[text] = generator("要學好人工智慧，你", max_length=100)
```

```python
''.join(text['generated_text'].split())
```

### 6. 生成歌詞

```python
generator = pipeline("text-generation", model="uer/gpt2-chinese-lyric")
```

```python
[lyric] = generator('来到一个陌生又熟悉的城市，', max_length=100)
```

```python
''.join(lyric['generated_text'].split())
```
